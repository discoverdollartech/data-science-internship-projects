```python
def get_subject(text):
    lines=re.split('\n',text)
    subjects=[]
    pat=re.compile('Subject:')
    for i in lines:
        if re.match(pat,i):
            subjects.append(i[9:])
    return subjects
```

```python
def get_first(data):
    
    l=data.split('From:')
    
    return l[0]
    
```

```python
REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|_\-\‚Äì<>;:]')
emails=re.compile('[\w.+-]+@[\w-]+\.[\w.-]+')
anchor=re.compile("<[^ ]*>")
cid=re.compile("cid[^ ]*")
sqbrac=re.compile('\[[^ ]*\]')
has=re.compile("#[0-9a-bA-B]*")

def clean_text(text):
    text=anchor.sub(" ",text)
    text=sqbrac.sub(" ",text)
    text=has.sub(" ",text)
    text=cid.sub(" ",text)
    text = REPLACE_BY_SPACE_RE.sub(' ', text)
    text=" ".join(text.split())
    return text
```

```python
import tensorflow as tf
import tensorflow_hub as hub
print("TF version: ", tf.__version__)
print("Hub version: ", hub.__version__)
```

```python
import tensorflow_text as text
```

```python
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string)
preprocessor = hub.KerasLayer(
    "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
encoder_inputs = preprocessor(text_input)
encoder = hub.KerasLayer(
    "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4",
    trainable=True)
outputs = encoder(encoder_inputs)
pooled_output = outputs["pooled_output"]      # [batch_size, 768].
sequence_output = outputs["sequence_output"]
```

```python
embedding_model = tf.keras.Model(text_input, pooled_output)
```

```python
sentences = tf.constant(["hi my name is sk"])
print(embedding_model(sentences))
```

```python
import numpy as np
data=[]
for i in range(100):
    print('processing ...',i)
    content=res['hits']['hits'][i]['_source']['content']
    content.encode("ascii", "replace")
    
    subjects=get_subject(content)
    
    if subjects:
        subject=subjects[0]
        sentences = tf.constant([subject])
        sub_vec=embedding_model(sentences)
    else:
        sub_vec=np.zeros((1, 768))
        
    text=get_first(content)
    
    
    if text:
        text_plain=clean_text(text)
        sentences=tf.constant([text_plain])
        text_vec=embedding_model(sentences)
    else:
        text_plain='None'
        text_vec=np.zeros((1, 768))
        
    
    data.append([subject,text_plain,sub_vec,text_vec])
    
    print('ending ....',i)
    
    print('----------------------------')
        
        

        
    
    
```

```python
query='increase in price'
sentence=tf.constant([query])
query_vector=embedding_model(sentence)
```

```python
query_vector.shape
```

# Cosine  similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
scores=[]
for row in data:
    
    sub_score=cosine_similarity(
        query_vector,
        row[2]
    )
    
    text_score=cosine_similarity(
        query_vector,
        row[3]
    )
    scores.append([row[0],row[1],sub_score,text_score])
```

```python
import csv  
header = ['subject', 'text', 'subject_similarity', 'text_similarity']


with open('scoress.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    
    writer.writerow(header)

    for i in scores:
        writer.writerow(i)
```
